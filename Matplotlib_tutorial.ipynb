```
# basic line plot
import matplotlib.pyplot as plt
import numpy as np
myarray = np.array([1,2,3,4])
plt.plot(myarray)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```
xarray = np.array([3,2,1])
yarray=np.array([6,5,4])
plt.scatter(xarray,yarray)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
