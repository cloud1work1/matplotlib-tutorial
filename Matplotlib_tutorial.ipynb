```
# basic line plot
import matplotlib.pyplot as plt
import numpy as np
myarray = np.array([1,2,3,4])
plt.plot(myarray)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```
xarray = np.array([3,2,1])
yarray=np.array([6,5,4])
plt.scatter(xarray,yarray)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```
# use of lineStyle
y = np.array([1,3,2,5,4])
plt.plot(y, lineStyle="dotted")
plt.show()
```

```
# use of lineStyle, color
y = np.array([1,3,2,5,4])
plt.plot(y, lineStyle="solid", color="indigo")
plt.show()
```

```
# use of lineStyle, color, linewidth
y = np.array([1,3,2,5,4])
plt.plot(y, lineStyle="--", color="r", lineWidth=2)
plt.show()
```

```
#use of multiplots
x=np.array([8.7,5.6,9.9,1,12.5])
y=np.array([4,5,2,1,7,5])
plt.plot(x)
plt.plot(y)
```

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#use of grid
y = [1,4, 9, 16, 25, 36,49]
x=[0,1,2,3,5,8,13]
x1=[1,1,2,6,24,120,720]
plt.plot(x,y)
plt.plot(x1,y)
plt.legend(labels=('tv','smartPhone'), loc='lower right')
plt.xlabel('Medium')
plt.ylabel('Sales')
plt.grid()
plt.title('Ad effects on sales')
plt.show()
```

```

#use of grid
y = [1,4, 9, 16, 25, 36,49]
x=[0,1,2,3,5,8,13]
x1=[1,1,2,6,24,120,720]
plt.plot(x,y, 'ys--')
plt.plot(x1,y, 'go--')
plt.legend(labels=('tv','smartPhone'), loc='lower right')
plt.xlabel('Medium')
plt.ylabel('Sales')
plt.grid(color='g', ls='-', lw=0.25)
plt.title('Ad effects on sales')
plt.show()
```
